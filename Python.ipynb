```
nota1 = float(input("Digite o 1 numero: "))
nota2 = float(input("Digite o 2 numero: "))
operacao = input("Digite a operacao (+,-,/,*)")

if operacao == "+":
  print("O resultado e ", nota1 + nota2)
elif operacao == "-":
  print("O resultado e ", nota1 - nota2)
elif operacao == "*":
  print("O resultado e ", nota1 * nota2)
elif operacao == "/":
  if nota2 != 0:
    print("O resultado e ", nota1 / nota2)
  else:
    print("Erro: Divisão por zero!")
else:
  print("Operação inválida.")
```

```
# Crie uma lista vazia chamada nomes.
# Adicione 3 nomes usando append() e mostre a lista usando f-string.
nomes = []
nomes.append("Ana")
nomes.append("Carlos")
nomes.append("Maria")
print(f"Lista de nomes: {nomes}")
```

```
# Crie uma lista com 3 cores.
# Adicione uma nova cor na posição 1 usando insert().
cores = ["azul", "verde", "vermelho"]
cores.insert(1, "amarelo")
print(f"Lista de cores atualizada: {cores}")
```

```
# Crie uma lista com 4 frutas.
# Remova a fruta "banana" usando remove().
frutas = ["maçã", "banana", "uva", "laranja"]
frutas.remove("banana")
print(f"Frutas após remoção: {frutas}")
```

```
# Crie uma lista com 5 números.
# Remova o último número usando pop().
numeros = [10, 20, 30, 40, 50]
removido = numeros.pop()
print(f"Número removido: {removido}")
print(f"Lista atual: {numeros}")
```

```
# Crie uma lista com 3 cidades; Substitua a segunda cidade por outra.
cidades = ["Belo Horizonte", "São Paulo", "Rio de Janeiro"]
cidades[1] = "Curitiba"
print(f"Lista atualizada: {cidades}")
```

```
# Crie duas listas de números; Una as duas listas usando extend().
lista1 = [1, 2, 3]
lista2 = [4, 5, 6]
lista1.extend(lista2)
print(f"Lista unida: {lista1}")
```

```
# Crie uma lista com nomes; Verifique se "Carlos" está na lista usando in.
nomes = ["Ana", "Carlos", "Maria"]
if "Carlos" in nomes:
    print(f"Carlos está na lista")
else:
    print(f"Carlos não está na lista")
```

```
# Crie uma lista com números desordenados.
# Ordene usando sort() e depois inverta com reverse().
numeros = [5, 2, 9, 1, 7]
numeros.sort()
print(f"Lista ordenada: {numeros}")
numeros.reverse()
print(f"Lista invertida: {numeros}")
```

```
# Crie uma lista com números repetidos. Mostre:
# quantas vezes o número aparece; a posição do primeiro elemento
numeros = [10, 20, 30, 20, 40, 20]
quantidade = numeros.count(20)
posicao = numeros.index(20)
print(f"O número 20 apareceu {quantidade} vezes")
print(f"Primeira posição do 20: {posicao}")
```
